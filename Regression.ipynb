```python
from sklearn.linear_model import LogisticRegression
X=[[30],[40],[50],[70]]
Y=[0,0,1,1]
model=LogisticRegression()
model.fit(X,Y)
print(model.predict([[60]]))
```

```python
from sklearn.linear_model import LinearRegression
X=[[30],[40],[50],[70]]
Y=[0,0,1,1]
model=LinearRegression()
model.fit(X,Y)
print(model.predict([[60]]))
```

```python
from sklearn.linear_model import LinearRegression
Hours = [[6], [9], [12], [15], [18]]
Tem = [[18], [22], [30], [35], [28]]
model = LinearRegression()
model.fit(Hours,Tem)
print(model.predict([[10]]))
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
Hours = np.array([[6], [9], [12], [15], [18]])
Tem = np.array([[18], [22], [30], [35], [28]])
model = LinearRegression()
model.fit(Hours,Tem)
hour = np.array([[14]])
predicted_temp = model.predict(hour)
print("Temperature at 14 hrs:", predicted_temp[0])
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
X = np.array([
    [1000, 2],
    [1500, 3],
    [2000, 4],
    [2500, 5]
])
y= np.array([1000, 1500, 2000, 2500])
model = LinearRegression()
model.fit(X, y)
new_house = np.array([[1200, 3]])
predicted_price = model.predict(new_house)
print("Predicted price for the new house:", predicted_price[0])
```
